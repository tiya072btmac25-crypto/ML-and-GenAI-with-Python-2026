```python
#QUESTION 1
#Create a function to print first 10 natural numbers.

def Natural_No(n):
  for i in range(1,n+1):
    print(i,end=" ")

Natural_No(10)
```

```python
#QUESTION 2
#Create a function to calculate sum of first N natural numbers.

n=int(input("Enter the no.: "))

def Sum(n):
  sum=0
  for i in range(1,n+1):
    sum=sum+i

  print(sum)

Sum(n)
```

```python
# QUESTION 3
# Create a function to reverse a number.

n=int(input("Enter the no.: "))

def reverse(n):
  rev=0
  digit=0

  while(n>0):
    digit=n%10
    rev=rev*10+digit
    n=n//10

  return rev

print("Reversed no.:",reverse(n))
```

```python
#QUESTION 4
#Create a function to count digits in a number.

n=int(input("Enter the no.: "))

def count(n):
  count=0
  while(n>0):
    n=n//10
    count=count+1

  return count

print("The no.of digits in the number : ",count(n))
```

```python
#QUESTION 5
#Create a function to check palindrome number.

n=int(input("Enter the no.: "))

def palindrome(n):
  rev=0
  digit=0
  temp=n

  while(n>0):
    digit=n%10
    rev=rev*10+digit
    n=n//10

  if rev==temp:
      print("The no. is a palindrome")
  else:
      print("The no. is not a palindrome")

palindrome(n)
```

```python
#QUESTION 6
#Create a function to generate Fibonacci series.

def fibonacci(n):
  a=0
  b=1

  for i in range(1,n+1):
    print(a,end=" ")
    c=a+b
    a=b
    b=c

fibonacci(10)
```

```python
#QUESTION 7
# Calculator Using Functions that contains the following features;
# - User selects operation
# - Program performs calculation
# - Display result

def add(x, y):
    """Adds two numbers."""
    return x + y

def subtract(x, y):
    """Subtracts the second number from the first."""
    return x - y

def multiply(x, y):
    """Multiplies two numbers."""
    return x * y

def divide(x, y):
    """Divides the first number by the second. Handles division by zero."""
    if y == 0:
        return "Error! Division by zero."
    return x / y

def calculator():
    print("--- Simple Python Calculator ---")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    # Get operation choice from user
    choice = input("Enter choice (1/2/3/4): ")

    # Check if choice is one of the valid options
    if choice in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter numerical values.")
            return

        # Perform calculation based on choice
        if choice == '1':
            print(f"Result: {num1} + {num2} = {add(num1, num2)}")
        elif choice == '2':
            print(f"Result: {num1} - {num2} = {subtract(num1, num2)}")
        elif choice == '3':
            print(f"Result: {num1} * {num2} = {multiply(num1, num2)}")
        elif choice == '4':
            print(f"Result: {num1} / {num2} = {divide(num1, num2)}")
    else:
        print("Invalid Input. Please select a valid operation.")

# Run the calculator
if __name__ == "__main__":
    calculator()

```
