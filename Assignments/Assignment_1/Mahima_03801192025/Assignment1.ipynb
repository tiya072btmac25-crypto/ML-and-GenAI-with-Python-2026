### Q1 - Find area of rectangle.

```python
length = int(input("Enter length of a rectangle: "))
breadth = int(input("Enter breadth of a rectangle: "))
area = length*breadth
print(area)
```

### Q2 - Find simple interest.

```python
p = int(input("Enter principal amount: "))
r = int(input("Enter rate of interest: "))
t = int(input("Enter time in years: "))
simple_interest = (p*r*t)/100
print(simple_interest)
```

### Q3 - Convert temperature from Celsius to Fahrenheit.

```python
temp_in_cel = int(input("Enter temperature in degree celcius: "))
temp_in_fah = (temp_in_cel * 1.8) + 32
print(temp_in_cel, "C = ", temp_in_fah, "F")
```

### Q4 - Calculate average of 3 numbers.

```python
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
n3 = int(input("Enter third number: "))
avg = (n1+n2+n3)/3
print("Average of three numbers: ", avg)
```

### Q5 - Find square and cube of a number.

```python
num = int(input("Enter a number: "))
square = num ** 2
cube = num ** 3
print("Square of ",num," is ", square)
print("Cube of ",num," is", cube)
```

### Q6 - Swap two numbers without third variable.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Before Swapping")
print("a = ", a)
print("b = ", b)
a, b = b, a
print("After swapping")
print("a = ", a)
print("b = ", b)
```

### Q7 - Create a Student Report Program that take student details using input(), Store marks in variables, 
### Calculate total and percentage , Use comments, Use proper indentation

```python
student_name = input("Enter student name: ")
student_marks = {
    "Math" : 91,
    'Physics' : 97,
    'Python' : 95
}
total_marks = student_marks["Math"] + student_marks["Physics"] + student_marks["Python"]
percentage = ((student_marks["Math"] + student_marks["Physics"] + student_marks["Python"]) / 300) * 100
print("Student Details")
print("Name : ",student_name)
print("Marks: ", student_marks)
print("Total Marks : ", total_marks)
print("Percentage : ", percentage, "%")
```
